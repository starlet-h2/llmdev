```python
# 成績のリスト
grades = [85, 90, 78, 92, 88]
#grades = [70, 75, 80, 85, 90]

# 成績の平均点を計算
average = sum(grades) / len(grades)
print(f"平均点: {average}")

# 成績の最高点を計算
highest = max(grades)
print(f"最高点: {highest}")

# 成績の最低点を計算
lowest = min(grades)
print(f"最低点: {lowest}")

# 成績が平均以上の学生数を計算
above_average_count = len([grade for grade in grades if grade >= average])
print(f"平均以上の学生数: {above_average_count}")
```
