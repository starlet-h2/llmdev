```python
import string

# 空のリストを作成
word_list = []

# キーがアルファベット小文字で、値が0の辞書を作成する
alphabet_dict = {letter: 0 for letter in string.ascii_lowercase}

# 英単語の入力をする
while True:
    word = input("英単語を入力してください: ")
    
    # 入力が空の場合はループを終了する
    if word == "":
        break
    
    # 英単語をリストに追加する
    word_list.append(word)

    # 英単語をアルファベットに分解して、辞書の値を更新する
    for letter in word:
        if letter in alphabet_dict:
            alphabet_dict[letter] += 1

# 英単語をアルファベット順にソートして表示する
sorted_words = sorted(word_list)
print("入力した英単語:", sorted_words)

# アルファベットの出現回数を表示する
# 1回以上出現したアルファベットのみ表示する
for letter, count in alphabet_dict.items():
    if count > 0:
        print(f"{letter}が{count}個ありました")
```
