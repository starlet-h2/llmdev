```python
# 元金（初期投資額）を入力
principal = float(input("元金を入力してください: "))

# 年利（パーセンテージ）を入力
annual_interest_rate = float(input("年利を入力してください: "))

# 複利計算(10年後の金額)
years = 10
while years > 0:
    principal += principal * (annual_interest_rate / 100)
    years -= 1

# 結果を表示
print(f"10年後の金額:  {principal:.2f}")
```
