```python
# 消費税率
tax_rate = 0.10

# 税抜価格を入力
tax_excluded_price = int(input("税抜価格を入力してください: "))

# 税込価格を計算(小数点以下は切り捨てとする)
tax_included_price = int(tax_excluded_price * (1 + tax_rate))

# 送料を計算
# 税込価格が2000円以上: 無料
# 税込価格が2000円未満: 350円
if tax_included_price >= 2000:
    shipping_fee = 0
    print("送料は無料です")
else:
    shipping_fee = 350
    print("送料は350円です")

# 合計金額を計算
total_price = tax_included_price + shipping_fee

# 送料込税込価格を表示
print(f"送料込税込価格は{total_price}円です")
```
