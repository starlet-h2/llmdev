```python
def validate_username(input_username):
    """ユーザー名のバリデーションを行う関数
    引数:
        input_username (str): ユーザー名の入力文字列
    戻り値:
        bool: バリデーションの結果（True: 有効, False: 無効）
    """
    # ユーザー名は5文字以上15文字以内かを確認
    if len(input_username) < 5 or len(input_username) > 15:
        print("エラー: ユーザー名は5文字以上15文字以内で入力してください")
        return False

    # ユーザー名が空白を含むかを確認
    if " " in input_username:
        print("エラー: ユーザー名に空白を含めることはできません")
        return False

    # ユーザー名が英数字（アルファベットと数字）のみかを確認
    if not input_username.isalnum():
        print("エラー: ユーザー名は英数字のみ使用できます")
        return False
    
    # すべての条件を満たしている場合はTrueを返す
    return True


# ユーザーからの入力を受け取る
username = input("ユーザー名を入力してください: ")

# バリデーション関数を呼び出し、結果を表示
if validate_username(username):
    print("ユーザー名は有効です")
else:
    print("不正なユーザー名です")

```
